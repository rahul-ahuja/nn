```python
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
from tensorflow.contrib.layers import flatten
import pickle
from sklearn.model_selection import train_test_split
import math
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import numpy as np
```

```python
with open('traffic-signs-data/train.p', 'rb') as f:
    data = pickle.load(f)
```

```python
X_train, X_val, y_train, y_val = train_test_split(data['features'], data['labels'], random_state=0, test_size=0.33, stratify=data['labels'])
```

```python
X_train = rgb2gray(X_train).astype('float32')
X_val = rgb2gray(X_val).astype('float32')
X_train = X_train / 255 - 0.5
X_val = X_val / 255 - 0.5
```

```python
plt.imshow(X_train[0], cmap='gray')
plt.show()
```

```python
image_size = 32
num_labels = 43
num_channels = 1 # grayscale

def reformat(dataset, labels):
  dataset = dataset.reshape(
    (-1, image_size, image_size, num_channels)).astype(np.float32)
  labels = (np.arange(num_labels) == labels[:,None]).astype(np.float32)
  return dataset, labels
train_dataset, train_labels = reformat(X_train,y_train)
val_dataset, val_labels = reformat(X_val,y_val)
print('Training set', train_dataset.shape, train_labels.shape)
print('Validation set', val_dataset.shape, val_labels.shape)
```

```python
### Train your model here.
### Feel free to use as many code cells as needed.
def accuracy(predictions, labels):
  return (100.0 * np.sum(np.argmax(predictions, 1) == np.argmax(labels, 1))
          / predictions.shape[0])
```

```python
batch_size = 128
patch_size = 5
depth = 16
num_hidden = 64
EPOCHS = 20
```

```python
tf_dataset = tf.placeholder(tf.float32, [None, image_size, image_size, num_channels])
tf_labels = tf.placeholder(tf.float32, [None, num_labels])
```

```python
def model(data):
    layer1_weights = tf.Variable(tf.truncated_normal(
          [patch_size, patch_size, num_channels, depth], stddev=0.01))
    layer1_biases = tf.Variable(tf.zeros([depth]))
    layer2_weights = tf.Variable(tf.truncated_normal(
          [patch_size, patch_size, depth, depth], stddev=0.01))
    layer2_biases = tf.Variable(tf.constant(1.0, shape=[depth]))
    conv = tf.nn.conv2d(data, layer1_weights, [1, 2, 2, 1], padding='SAME')
    hidden = tf.nn.relu(conv + layer1_biases)
    conv = tf.nn.conv2d(hidden, layer2_weights, [1, 2, 2, 1], padding='SAME')
    hidden = tf.nn.relu(conv + layer2_biases)
    #shape = hidden.get_shape().as_list()
    fc1 = flatten(hidden)
    #reshape = tf.reshape(hidden, [shape[0], shape[1] * shape[2] * shape[3]])
    fc1_shape = (fc1.get_shape().as_list()[-1], num_labels)
    fc1_W = tf.Variable(tf.truncated_normal(shape=(fc1_shape), stddev=0.01))
    fc1_b = tf.Variable(tf.zeros(num_labels))
    return tf.matmul(fc1, fc1_W) + fc1_b
```

```python
  # Training computation.
logits = model(tf_dataset)
loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits, tf_labels))
    
  # Optimizer.
optimizer = tf.train.GradientDescentOptimizer(0.05).minimize(loss)
# Predictions for the training, validation, and test data.
prediction = tf.nn.softmax(logits)
```
